```python
import numpy
import pandas
```

```python
import tensorflow.contrib.keras as keras # this is important

from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense
from tensorflow.python.keras.wrappers.scikit_learn import KerasClassifier
from tensorflow.python.keras.utils import np_utils
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
```

```python
from sklearn import datasets
import numpy as np

iris = datasets.load_iris()
X = iris.data
Y = iris.target
# convert integers to dummy variables (i.e. one hot encoded)
dummy_y = np_utils.to_categorical(Y)
```

```python
# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
```

## Define NN

```python
# define baseline model
def baseline_model():
    # create model
    model = keras.models.Sequential()
    model.add(Dense(8, input_dim=4, activation='relu'))
    model.add(Dense(3, activation='softmax'))
    # Compile model
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
## SKLearn wrapper
nn = KerasClassifier(build_fn=baseline_model, epochs=200, batch_size=5, verbose=0)
```

```python
nn.fit(X, dummy_y)
```

```python
print('Training accuracy:', nn.score(X, dummy_y))
```

## K-fold CV

```python
kfold = KFold(n_splits=5, shuffle=True, random_state=seed)
```

```python
results = cross_val_score(nn, X, dummy_y, cv=kfold)
print("Baseline: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```
